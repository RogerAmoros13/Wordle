```python
import pandas as pd
import numpy as np
import unidecode
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('words.csv')
data.head()
```

```python
data['count'] = data['words'].apply(lambda x: len(x))
data.drop(data[data['count']!=5].index, inplace=True)
data.head()
```

```python
data['words'] = data['words'].apply(lambda x: unidecode.unidecode(x))

data.head()
```

```python
data = data.groupby('words').count().reset_index()
data['words'] = data['words'].str.lower()
data = data.drop(columns=['count'])
data.head()
```

```python
data['letter'] = data['words'].apply(lambda x: ''.join(set(x)))
data['diversity'] = data['letter'].apply(lambda x: len(x) / 5)
data.tail()
```

```python
letras = {}
for i in range(len(data)):
    for let in data.loc[i, 'letter']:
        if let not in letras.keys():
            letras[let] = [0]
        letras[let][0] += 1

df_letras = pd.DataFrame(letras)
df_letras = df_letras.transpose().reset_index()
df_letras.rename(columns={'index': 'lletra', 0: 'freq_abs'}, inplace=True)
df_letras = df_letras.sort_values(by='freq_abs', ascending=False)
```

```python
df_letras['freq_rel'] = df_letras['freq_abs'] / data.shape[0]
df_letras.head()
```

```python
def calculate_result(name):
    i = 0
    for let in name:
        if i == 0:
            mult = df_letras[df_letras['lletra']==let].iloc[0]['freq_rel']
            
            i += 1
        else:
            mult = mult + df_letras[df_letras['lletra']==let].iloc[0]['freq_rel']
    return mult / 4
data['ranking'] = data['letter'].map(calculate_result)
data.head()
```

```python
data = data.sort_values(by='ranking', ascending=False)
data.drop(index=[19], inplace=True)
data.tail()
```

```python
data.to_csv('words_def.csv')
```
